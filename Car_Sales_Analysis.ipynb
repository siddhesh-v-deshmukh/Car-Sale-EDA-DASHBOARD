```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('car_prices.csv')
```

```python
df.drop('transmission', axis=1, inplace=True)
```

```python
df.dropna(subset=['make', 'model', 'trim', 'body'], inplace=True)
```

```python
df['condition'].fillna(df['condition'].median(), inplace=True)
```

```python
df['odometer'].fillna(df['odometer'].median(), inplace=True)
```

```python
df['color'].fillna(df['color'].mode()[0], inplace=True)
df['interior'].fillna(df['interior'].mode()[0], inplace=True)
```

```python
df.dropna(subset=['mmr', 'sellingprice', 'saledate'], inplace=True)
```

```python
state_seller_count=df.groupby('state')['seller'].nunique().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(12,8))
ax=state_seller_count.plot(kind='bar',color='mediumseagreen')

for i,(state,count) in enumerate(state_seller_count.items()):
    ax.text(i, count + 0.5, str(count), ha='center', va='bottom', fontsize=9)

plt.title('state vise seller')
plt.xlabel('state')
plt.ylabel('count of seller')
plt.xticks(rotation=45)
plt.show()
```

```python
df['body'] = df['body'].str.lower().str.strip()
```

```python
avg_body_per_price=df.groupby('body')['sellingprice'].mean().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x=avg_body_per_price.index, y=avg_body_per_price.values, palette='coolwarm')
plt.title('Average price by body type')
plt.xlabel('Body Type')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
for i, value in enumerate(avg_body_per_price.values):
    plt.text(i, value + 500, f'{value:,.0f}', ha='center', va='bottom', fontsize=10)
plt.tight_layout()
plt.show
```

```python
top_models = df['model'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(
    x=top_models.values,
    y=top_models.index,
    palette='viridis',
    hue=top_models.index,      # explicitly setting hue
    legend=False               # disabling legend since hue is just for color
)
plt.title('Top 10 Car Models by Sales')
plt.xlabel('Number of Cars Sold')
plt.ylabel('Car Model')
plt.tight_layout()
plt.show()
```

```python
top_color = df['color'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(
    x=top_color.values,
    y=top_color.index,
    palette='viridis',
    hue=top_color.index,      # explicitly setting hue
    legend=False               # disabling legend since hue is just for color
)
plt.title('Top 10 Car Color')
plt.xlabel('Number of Cars')
plt.ylabel('Car Color')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
avg_price_by_year = df.groupby('year')['sellingprice'].mean().reset_index()
sns.lineplot(x='year', y='sellingprice', data=df, estimator='mean', color='green', marker='o')
for i, row in avg_price_by_year.iterrows():
    plt.text(row['year'], row['sellingprice'] + 500, f'{row["sellingprice"]:,.0f}', 
             ha='center', va='bottom', fontsize=9)
plt.title('Average Car Price by Sale Year')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
state_condition_avg = df.groupby('state')['condition'].mean().sort_values(ascending=False).head(20)
```

```python
plt.figure(figsize=(10,6))
plt.figure(figsize=(12, 8))
sns.barplot(x=state_condition_avg.values, y=state_condition_avg.index,palette='coolwarm')
plt.title('Average Car Condition by State')
plt.xlabel('Average Condition (1 = Worst, 49 = Best)')
plt.ylabel('State')
for i, (value, state) in enumerate(zip(state_condition_avg.values, state_condition_avg.index)):
    plt.text(value + 0.3, i, str(int(value)), va='center', fontsize=9, color='black')
plt.tight_layout()
plt.show()
```

```python
# 1. Remove timezone and anything in parentheses
df['clean_saledate'] = df['saledate'].str.replace(r'GMT.*|\(.*\)', '', regex=True).str.strip()

# 2. Parse the cleaned column to datetime
df['clean_saledate'] = pd.to_datetime(df['clean_saledate'], errors='coerce')
```

```python
df.head()
```

```python
df['clean_saledate'] = pd.to_datetime(df['clean_saledate'], errors='coerce')

df['year_month'] = df['clean_saledate'].dt.to_period('M')  # e.g., 2020-03

monthly_sales = df['year_month'].value_counts().sort_index()

monthly_sales.plot(kind='line', figsize=(12, 6), marker='o')
plt.title('Monthly Car Sales Trend')
plt.xlabel('Month')
plt.ylabel('Number of Cars Sold')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
bins = [0, 10, 20, 30, 40, 50]
labels = ['0–10', '11–20', '21–30', '31–40', '41–50']
df['condition_range'] = pd.cut(df['condition'], bins=bins, labels=labels, include_lowest=True)


avg_price_by_condition = df.groupby('condition_range')['sellingprice'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=avg_price_by_condition, x='condition_range', y='sellingprice', palette='viridis')

plt.title('Average Car Price by Condition Range')
plt.xlabel('Condition Range')
plt.ylabel('Average Selling Price')
plt.tight_layout()
plt.show()
```
